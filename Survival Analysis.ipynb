# **Uji Log-Rank**

```python
from lifelines.statistics import logrank_test
import pandas as pd
from itertools import combinations

# Daftar kategori Contract
contract_columns = ['Contract_Month-to-month', 'Contract_One year', 'Contract_Two year']
contract_labels = [col.replace('Contract_', '') for col in contract_columns]

# Simpan hasil uji log-rank
results = []

# Kombinasikan setiap pasangan Contract
for (col1, label1), (col2, label2) in combinations(zip(contract_columns, contract_labels), 2):
    group1 = data_labeled[data_labeled[col1] == 1]
    group2 = data_labeled[data_labeled[col2] == 1]

    test_result = logrank_test(
        group1['tenure'], group2['tenure'],
        event_observed_A=group1['Churn'],
        event_observed_B=group2['Churn']
    )

    results.append({
        'Group 1': label1,
        'Group 2': label2,
        'Test Statistics': round(test_result.test_statistic,2),
        'p-value': test_result.p_value
    })

# Masukkan ke DataFrame
logrank_df = pd.DataFrame(results)
logrank_df['Significant'] = logrank_df['p-value'] < 0.05
logrank_df
```

Tujuan hasil uji log-rank adalah untuk membandingkan kurva survival antara dua atau lebih kelompok, yaitu jenis kontrak dan menguji apakah terdapat perbedaan yang signifikan secara statistik dalam probabilitas bertahan (survival) antar kelompok tersebut.

### **Interpretasi**
Berdasarkan output di atas diperoleh hasil uji log-rank yang menunjukkan bahwa terdapat perbedaan yang signifikan dalam kelangsungan pelanggan antar jenis kontrak. Pelanggan dengan kontrak month-to-month memiliki kecenderungan churn atau berhenti berlangganan yang lebih tinggi dibanding pelanggan dengan kontrak one year maupun two year. Selain itu, perbedaan antara kontrak one year dan two year juga terlihat cukup signifikan yang berarti bahwa semakin panjang durasi kontrak, maka semakin tinggi peluang pelanggan untuk tetap bertahan atau no churn.

# **Modelling**

## **Model Cox Proportional Hazards (CoxPH)**

**Alasan:**
- **Tipe Data:** Data churn berisi informasi waktu hingga pelanggan berhenti berlangganan (tenure) dan status churn (apakah pelanggan sudah churn atau belum)
- **Censoring:** Tidak semua pelanggan sudah churn, tetapi sebagian masih aktif (right-censored)
- **Multivariat:** CoxPH memungkinkan analisis pengaruh banyak variabel (fitur) terhadap risiko churn secara simultan
- **Tidak Perlu Asumsi Distribusi Survival:** CoxPH tidak mengasumsikan bentuk khusus distribusi waktu survival (non-parametrik pada baseline hazard)

**Tujuan:**
- **Memprediksi Risiko Churn:** Mengestimasi seberapa besar risiko pelanggan untuk churn pada waktu tertentu berdasarkan karakteristik/faktor yang dimiliki
- **Mengidentifikasi Faktor Risiko:** Mengetahui fitur/variabel apa saja yang paling berpengaruh terhadap kemungkinan pelanggan churn
- **Segmentasi Risiko:** Membagi pelanggan ke dalam kelompok risiko tinggi/rendah untuk strategi retensi yang lebih efektif
- **Prediksi Survival Probability:** Menghitung probabilitas bertahan pelanggan pada waktu tertentu

```python
# Drop one column from each one-hot encoded group to avoid perfect multicollinearity
drop_cols = [
    'InternetService_No',      # drop one from InternetService
    'Contract_Two year',       # drop one from Contract
    'PaymentMethod_Mailed check' # drop one from PaymentMethod
]

all_encoded_features = [
    col for col in data_labeled.columns if col not in ['tenure', 'Churn'] and col not in drop_cols
]

features_for_cox = [col for col in all_encoded_features if col not in drop_cols]

# Model Cox Proportional Hazards
from lifelines import CoxPHFitter
cph = CoxPHFitter()
cph.fit(data_labeled[features_for_cox + ['tenure', 'Churn']], duration_col='tenure', event_col='Churn')
```

## **Backward Elimination**

Backward elimination adalah metode seleksi fitur secara otomatis yang dilakukan dengan cara **menghapus fitur yang paling tidak signifikan** berdasarkan p-value secara bertahap dari model hingga hanya tersisa fitur-fitur yang signifikan.

```python
from lifelines import CoxPHFitter

# Mapping fitur kategorik ke prefix one-hot encoding
categorical_prefixes = [
    'InternetService_', 'Contract_', 'PaymentMethod_'
]

def get_related_onehot_features(feature, all_features):
    # Jika fitur adalah one-hot hasil kategorik, hapus semua fitur dengan prefix yang sama
    for prefix in categorical_prefixes:
        if feature.startswith(prefix):
            return [f for f in all_features if f.startswith(prefix)]
    return [feature]

def backward_elimination(data, features, duration_col, event_col, p_threshold=0.05):
    features = features.copy()
    while True:
        cph = CoxPHFitter()
        cph.fit(data[features + [duration_col, event_col]], duration_col=duration_col, event_col=event_col)
        p_values = cph.summary['p']
        max_p = p_values.max()
        if max_p > p_threshold:
            remove_feature = p_values.idxmax()
            related_features = get_related_onehot_features(remove_feature, features)
            for f in related_features:
                if f in features:
                    features.remove(f)
                    print(f"Remove: {f} (p={p_values.get(f, max_p):.4f})")
        else:
            break
    return features, cph

selected_features, cph_selected = backward_elimination(
    data_labeled, features_for_cox, duration_col='tenure', event_col='Churn', p_threshold=0.05
)

cph_selected.print_summary()
```

## **Pemilihan Penalizer Optiamal Berdasarkan Concordance Index**

```python
from lifelines import CoxPHFitter
from lifelines.utils import concordance_index

# Range penalizer yang akan diuji
penalizers = [0.001, 0.01, 0.05, 0.1, 0.5, 1]
ratios = [0, 1, 0.5]  # 0: Ridge, 1: Lasso, 0.5: Elastic Net

results = []

for l1_ratio in ratios:
    best_cindex = -1
    best_penalizer = None
    for pen in penalizers:
        cph = CoxPHFitter(penalizer=pen, l1_ratio=l1_ratio)
        cph.fit(data_labeled[features_for_cox + ['tenure', 'Churn']], duration_col='tenure', event_col='Churn')
        cindex = concordance_index(
            data_labeled['tenure'],
            -cph.predict_partial_hazard(data_labeled[features_for_cox]),
            data_labeled['Churn']
        )
        results.append({
            'Model': 'Ridge' if l1_ratio == 0 else 'Lasso' if l1_ratio == 1 else 'Elastic Net',
            'l1_ratio': l1_ratio,
            'penalizer': pen,
            'concordance_index': cindex
        })
        if cindex > best_cindex:
            best_cindex = cindex
            best_penalizer = pen
    print(f"Best penalizer for {'Ridge' if l1_ratio == 0 else 'Lasso' if l1_ratio == 1 else 'Elastic Net'} (l1_ratio={l1_ratio}): {best_penalizer} with concordance_index={best_cindex:.4f}")

# Hasil lengkap dalam DataFrame
pd.DataFrame(results)
```

# **Evaluasi Model Survival Analysis**

## **Concordance Index**
Concordance index (C-index) mengukur seberapa baik model survival analysis (misal CoxPH) memprediksi urutan relatif waktu churn pelanggan.

Pada konteks dataset churn:
- Untuk dua pelanggan $(i, j)$:
    * Jika pelanggan $i$ churn lebih cepat ($T_i < T_j$) dan model juga memprediksi risiko churn $i$ lebih tinggi ($\hat{h}(x_i) > \hat{h}(x_j)$), maka pasangan tersebut **concordant**.
    * Jika prediksi model berlawanan dengan kenyataan, disebut **discordant**.

C-index dirumuskan sebagai:
$$
\text{C-index} = \frac{\# \text{concordant pairs}}{\# \text{comparable pairs}}
$$

"Comparable pairs" hanya dihitung untuk pasangan di mana pelanggan yang churn lebih awal **tidak disensor** (benar-benar churn, bukan masih aktif).

Interpretasi nilai C-index:
- $0.5$: prediksi model acak (seperti lempar koin)
- $> 0.7$: model cukup baik dalam mengurutkan risiko churn
- $1.0$: model sempurna (semua urutan prediksi benar)
